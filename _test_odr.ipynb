```python
import numpy as np
import matplotlib.pyplot as plt

import util
```

```python
n = 10
x = np.linspace(0.0, 10.0, n)
xerr = np.random.normal(0, 2.5, n)

y = np.linspace(0., 30., n)
yerr = np.random.normal(0, 2.5, n)

x = x + xerr
y = y + yerr

plt.plot(x, y, 'o', label='data')
plt.plot(x-xerr, y-yerr, '-', label='truth')

linreg_odr = util.linreg_odr(x, y, xerr=1./xerr, yerr=1./yerr)
plt.plot(linreg_odr.xhat, linreg_odr.yhat, '-', label='ODR')

plt.legend()

print(linreg_odr.pval)
```
